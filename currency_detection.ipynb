```python
import re
import os

# Ensure the necessary paths exist
input_file = '/content/models/research/object_detection/packages/tf2/setup.py'
output_file = '/content/models/research/setup.py'

# Step 1: Modify setup.py to target tf-models-official==2.8.0
if os.path.exists(input_file):
    with open(input_file, 'r') as f:
        setup_content = f.read()

    # Replace the required dependency version
    modified_content = re.sub(r'tf-models-official>=2\.5\.1', 'tf-models-official==2.8.0', setup_content)

    with open(output_file, 'w') as f:
        f.write(modified_content)

    print("setup.py modified successfully.")
else:
    print(f"File not found: {input_file}")

# Step 2: Navigate to the research directory and install the package
!cd /content/models/research/ && pip install .
!git clone --depth 1 https://github.com/tensorflow/models.git
!pip list | grep tf-models-official
```
